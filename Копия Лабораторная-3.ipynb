## Задание 1
Для каждого артиста вывести все жанры, которые есть в его песнях, и для каждого жанра вывести наиболее продолжительную песню

```python
%%sql
SELECT
       ArtistName
       ,GenreName
       ,TrackName
FROM 
      (SELECT ar.Name      AS ArtistName
              ,g.Name      AS GenreName
              ,tracks.Name AS TrackName
              ,dense_rank() OVER (PARTITION BY ar.Name, g.Name ORDER BY milliseconds DESC ) AS dr
       FROM tracks

       LEFT JOIN ( SELECT * FROM albums  ) al USING (AlbumId)
                 ,(SELECT * FROM artists ) ar USING (ArtistId)
                 ,(SELECT * FROM genres  ) g  USING (GenreId)
      ) 
WHERE dr = 1
```

## Задание 2 
Для каждой компании вывести трек, за который компания заплатила наибольшую сумму, вывести жанр данного трека и его альбом

```python
%%sql
SELECT 
      CompanyName
      ,TrackName
      ,GenreName
      ,AlbumName
      ,SUP ---Нужн максимальное число денег, отданное за трек в сумме
      ,MUP ---или нужен просто самый дорогой купленный трек?
FROM
    (SELECT
            c.Company AS CompanyName
            ,t.Name   AS TrackName
            ,g.Name   AS GenreName
            ,a.Title  AS AlbumName
            ,SUM(t.UnitPrice) as SUP
            ,MAX(t.UnitPrice) as MUP
            ,row_number() OVER (PARTITION BY c.Company ORDER BY SUM(t.UnitPrice) DESC) as rn
            ---,*
    FROM
          tracks t

    LEFT JOIN  (SELECT *  FROM albums  )   a USING (AlbumId)
              ,(SELECT *  FROM genres  )   g USING (GenreId)
              ,(SELECT *  FROM invoice_items ii ,invoices i                      ---объединил invoices и invoice_items
                          WHERE ii.InvoiceId = i.InvoiceId ) al USING (TrackId)
              ,(SELECT * FROM customers )  c USING (CustomerId)

    WHERE not (Company is  null)
    
    GROUP BY Company
    )
WHERE rn = 1
```

## Задание 3
Для каждого плейлиста вывести трек с наибольшей продолжительностью, трек с наименьшей стоимостью (если у треков одинаковая стоимость, то вывести тот, который весит меньше всех), и жанр данного плейлиста (жанром плейлиста называется тот жанр, треков которого больше всех в плейлисте).

```python
%%sql
WITH NewTrackList AS 
(SELECT 
      *
FROM
      (SELECT DISTINCT
          pl.Name as PlayListName
          ,TrackName
          ,GenreName
          ,PlayListId
          
          ,first_value(TrackName) over (partition by PlayListId ORDER BY milliseconds DESC) LongestTrack
          ,first_value(TrackName) over (partition by PlayListId ORDER BY UnitPrice ASC ,    
                                                                             Bytes ASC)     CheapesOrLightest
          ,rank() over (partition by PlayListId ORDER BY GenreId DESC) rk                 ---смотрим самый большой ранк в плейлисте
          

          ---For_Test
          --- ,dense_rank() over (partition by PlayListId ORDER BY milliseconds DESC) dr  ---наибольшая продолжительность
          --- ,dense_rank() over (partition by PlayListId ORDER BY UnitPrice ASC , 
          ---                                                          Bytes ASC)    dr2  ---наименьшая стоимость (или размер)
      FROM
          playlists pl

      LEFT JOIN  (SELECT * FROM playlist_track)                    pt USING (PlaylistId)
                 ,(SELECT Name as TrackName ,* FROM tracks)        t  USING (TrackId)
                 ,(SELECT * ,genres.Name as GenreName FROM genres) g  USING (GenreId))

GROUP BY PlayListId  
         ,PlayListName
)


SELECT   ---тут можно без оператора WITH всё сделать, но я слишком поздно это заметил
     PlayListName
     ,LongestTrack      ---наибольшая продолжительность
     ,CheapesOrLightest ---наименьшая стоимость (или размер)
     ,GenreName
FROM 
      NewTrackList
```
